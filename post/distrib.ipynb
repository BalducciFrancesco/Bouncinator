```python
import matplotlib.pyplot as plt
import pandas as pd
import glob
import os

for csv in glob.glob(os.path.join(os.getcwd(), "*.csv")):
    df = pd.read_csv(csv)
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    plt.suptitle(f'Score Distributions for {os.path.basename(csv)}')
    for i, score in enumerate(['Score1', 'Score2']):
        counts = df[score].value_counts().sort_index()
        mean = df[score].mean()
        std = df[score].std()
        counts.plot(kind='bar', ax=axes[i], color=f'C{i}')
        axes[i].set_title(f'{score} (Mean: {mean:.2f}, Std: {std:.2f})')
        axes[i].set_ylabel('Count')
        axes[i].set_xticklabels(counts.index, rotation=0)
        axes[i].grid(axis='y')
        axes[i].set_xlabel('Score')
        for idx, value in enumerate(counts.values):
            axes[i].text(idx, value + 1, str(value), ha='center', va='bottom', fontsize=10)
    plt.tight_layout()
    plt.show()
```
